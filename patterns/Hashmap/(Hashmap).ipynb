# Hashmap 

Hashing problems are usually one of these:

1. Lookup / Membership
2. Frequency Tracking
3. Index Mapping
4. Pair-Sum / Complement Storage
5. State Compression
6. O(1) Data Structure Design

Category 1: Membership & Presence Tracking
Core Idea

Use HashSet/HashMap for O(1) existence checks.

Recognition Signals: 
```python
lookup
existence
visited
consecutive elements
duplicates
```

| Order | Problem                               | Label   | Core Idea                        |
| ----- | ------------------------------------- | ------- | -------------------------------- |
| 1     | Contains Duplicate (LC 217)           | MUST DO | Presence check                   |
| 2     | Longest Consecutive Sequence (LC 128) | MUST DO | Expand only from sequence starts |


Category 2: Frequency-State Hashing

Core Idea

HashMap stores counts/frequencies.

Recognition Signals:

```python
most frequent
counts
anagrams
frequency matching
```

| Order | Problem                               | Label    |
| ----- | ------------------------------------- | -------- |
| 1     | Valid Anagram (LC 242)                | MUST DO  |
| 2     | Ransom Note (LC 383)                  | Variant  |
| 3     | Top K Frequent Elements (LC 347)      | MUST DO  |
| 4     | Top K Frequent Words (LC 692)         | Variant  |
| 5     | Sort Characters By Frequency (LC 451) | Variant  |
| 6     | Reorganize String (LC 767)            | Advanced |

Core Idea

Hashmap is infrastructure.

```python
O(1) insert
O(1) delete
O(1) lookup
custom data structure
```
