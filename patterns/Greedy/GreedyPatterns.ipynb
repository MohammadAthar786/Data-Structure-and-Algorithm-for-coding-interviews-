# Greedy Algorithms Pattern Sheet

## Core Philosophy

Greedy algorithms make the best local decision at the current step while preserving global optimality.

The most important question is NOT:

```text
What is the locally best choice?
```

The most important question is:

```text
Why is this local choice safe?
```

Every greedy problem should be studied with:

1. Greedy decision
2. Why it is safe
3. Counterexample if greedy fails

---

# Level 0: Greedy Recognition Test

Greedy is likely applicable when:

* making a sequence of decisions
* maximizing or minimizing something
* sorting helps expose structure
* local choices appear irreversible
* problem asks for optimal schedule, allocation, coverage, or selection

Typical signals:

* earliest finish
* smallest/largest first
* maximum reach
* minimum resources
* best profit
* interval scheduling
* resource allocation

# Category 1: Foundational Greedy

## Core Idea

Learn why local choices can become globally optimal.

---

### Problem Ladder

| Order | Problem                               | Label   | Greedy Decision                    |
| ----- | ------------------------------------- | ------- | ---------------------------------- |
| 1     | Activity Selection                    | MUST DO | Choose earliest finishing activity |
| 2     | Minimum Coins (Canonical Coin System) | MUST DO | Take largest denomination first    |

---

## Invariant

```text
Current decision leaves maximum flexibility
for future decisions.
```

---

## Important Warning

Greedy coin selection only works for specific coin systems.

Example:

```text
Coins = [1,3,4]
Amount = 6

Greedy:
4 + 1 + 1

Optimal:
3 + 3
```

Greedy is not universally valid.
