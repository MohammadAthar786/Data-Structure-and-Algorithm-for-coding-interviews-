# Greedy Algorithms: Proof-Driven Pattern Sheet

## Greedy Recognition Test

A greedy solution needs more than a locally attractive choice. State why replacing any different choice with the greedy choice cannot make the final answer worse: an exchange argument, dominance argument, or irreversible frontier invariant.

`Added bridge` marks an interview-worthy addition that completes a ladder.

## 1. Earliest-Finish Interval Selection

**Invariant:** Among accepted schedules of the same size, keeping the interval that finishes earliest leaves at least as much room for every future interval.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Activity Selection | Learn earliest-finish exchange argument | Must do |
| 2 | Non-overlapping Intervals (LC 435) | Count removals instead of accepted intervals | Must do |
| 3 | Minimum Arrows to Burst Balloons (LC 452) | One selected endpoint covers an overlap group | Must do |

## 2. Sorted Interval Scanning

**Invariant:** After sorting by start, all intervals before the current one are already merged or proven compatible.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Meeting Rooms I (LC 252) | Detect whether consecutive intervals overlap | Must do |
| 2 | Merge Intervals (LC 56) | Extend the active merged interval | Must do |
