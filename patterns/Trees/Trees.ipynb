# Trees: Invariant-Based Interview Sheet

## Recognition Rule

First decide what one recursive call returns: a node, height, boolean, path contribution, or multiple states. For BFS, state what one queue level or coordinate represents.

`Added bridge` marks a high-value addition used to complete a ladder.

## 1. Recursive DFS Traversals

**Invariant:** The position of `visit(node)` relative to the left and right recursive calls determines preorder, inorder, or postorder.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Binary Tree Preorder Traversal (LC 144) | Process before children | Must do |
| 2 | Binary Tree Inorder Traversal (LC 94) | Process between children | Must do |
| 3 | Binary Tree Postorder Traversal (LC 145) | Process after children | Must do |

## 2. Level-Order BFS

**Invariant:** At the start of an outer iteration, the queue contains exactly one tree level.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Binary Tree Level Order Traversal (LC 102) | Process one queue level | Must do |
| 2 | Reverse Level Order Traversal (LC 107) | Reverse level collection order | Recommended |
| 3 | Zigzag Level Order Traversal (LC 103) | Alternate write direction by level | Must do |
