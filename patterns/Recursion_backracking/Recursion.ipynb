##  1. Inclusion / Exclusion Pattern (Pick / Not Pick)

** Core Intuition:** At each index, you have two choices: Include the current element (use it once) or Exclude the current element (move to next).

** Template:**
```cpp
int solve(int i, vector<int>& arr, int target) {
    if (i >= arr.size()) return (target == 0) ? 1 : 0;
    
    int pick = solve(i + 1, arr, target - arr[i]);    // Include current
    int notPick = solve(i + 1, arr, target);          // Exclude current
    
    return pick + notPick;
}
```

##  2. Index Skipping Pattern

** Core Idea:** At each index i, you can either pick the element (skip i+1) or not pick (go to i+1). Avoids selecting adjacent elements.

** Template:**
```cpp
int solve(int i, vector<int>& arr) {
    if (i >= arr.size()) return 0;
    
    int pick = arr[i] + solve(i + 2, arr);     // Skip next after picking
    int notPick = solve(i + 1, arr);           // Go to next normally
    
    return max(pick, notPick);
}
```

##  3. Pick / Not Pick with Infinite Supply

** Core Idea:** Element can be chosen multiple times. On "pick", stay at same index; on "not pick", move to next.

** Template:**
```cpp
int solve(int i, vector<int>& arr, int target) {
    if (target == 0) return 1;
    if (i >= arr.size() || target < 0) return 0;
    
    int pick = solve(i, arr, target - arr[i]);      // Stay at same index
    int notPick = solve(i + 1, arr, target);        // Move to next
    
    return pick + notPick;
}
```

##  4. Permutation Generation Pattern

** Core Idea:** At each level, pick one unused element, recurse deeper, then backtrack.

** Template (Swap-based):**
```cpp
void permute(vector<int>& arr, int index) {
    if (index == arr.size()) {
        result.push_back(arr);
        return;
    }
    for (int i = index; i < arr.size(); i++) {
        swap(arr[i], arr[index]);
        permute(arr, index + 1);
        swap(arr[i], arr[index]); // backtrack
    }
}
```

##  5. Fixed Choices at Each State (Tree-shaped)

** Core Idea:** Each recursion level has a fixed set of choices (directions, operators, letters). Tree fans out based on available options.

** Template:**
```cpp
void solve(int i, string& current, vector<string>& result) {
    if (i == target_len) {
        result.push_back(current);
        return;
    }
    
    for (char choice : fixed_choices) {
        current.push_back(choice);
        solve(i + 1, current, result);
        current.pop_back();
    }
}
```

## Level 2: Grid/Direction Based

###  Sub type : Grid-based DFS Recursion (Matrix Backtracking)

** Core Idea:** Explore grid cells recursively in 4 directions while tracking visited cells to avoid cycles.

** Template:**
```cpp
void dfs(int row, int col, vector<vector<char>>& grid, vector<vector<bool>>& visited) {
    int m = grid.size(), n = grid[0].size();
    
    if (row < 0 || col < 0 || row >= m || col >= n || visited[row][col])
        return;
    
    visited[row][col] = true;
    
    // 4 directions
    int dx[] = {-1, 0, 1, 0};
    int dy[] = {0, 1, 0, -1};
    
    for (int d = 0; d < 4; ++d) {
        dfs(row + dx[d], col + dy[d], grid, visited);
    }
    
    visited[row][col] = false; // backtrack if needed
}
```

##  7. Unbounded Loop-based Recursion

** Core Idea:** Loop through remaining options at each level and recursively explore from each choice.

** Template:**
```cpp
void backtrack(vector<int>& path, int start) {
    // Process current path if needed
    
    for (int i = start; i < choices.size(); i++) {
        if (isValid(choices[i], path)) {
            path.push_back(choices[i]);
            backtrack(path, i); // or i+1 based on reuse
            path.pop_back();
        }
    }
}
```

##  8. Multiple Recursive Branches

** Core Idea:** One recursive call spawns multiple other recursive calls, combining their results.

** Template:**
```cpp
int solve(int n) {
    if (n <= 1) return 1;
    
    int ans = 0;
    for (int option : options) {
        if (valid(option, n)) {
            ans += solve(n - option);
        }
    }
    return ans;
}
```

##  9. Recursive Partitioning

** Core Idea:** Recursively split input into valid partitions and explore all combinations.

** Template:**
```cpp
void solve(string& s, int idx, vector<string>& path) {
    if (idx == s.length()) {
        result.push_back(path);
        return;
    }
    
    for (int i = idx; i < s.length(); ++i) {
        string part = s.substr(idx, i - idx + 1);
        
        if (isValid(part)) {
            path.push_back(part);
            solve(s, i + 1, path);
            path.pop_back();
        }
    }
}
```

##  10. Divide-and-Conquer Recursion

** Core Idea:** Divide problem into independent subproblems, solve recursively, and combine results.

** Template:**
```cpp
int solve(vector<int>& arr, int left, int right) {
    if (left == right) {
        return arr[left]; // base case
    }
    
    int mid = (left + right) / 2;
    int leftAns = solve(arr, left, mid);
    int rightAns = solve(arr, mid + 1, right);
    
    return combine(leftAns, rightAns);
}
```
