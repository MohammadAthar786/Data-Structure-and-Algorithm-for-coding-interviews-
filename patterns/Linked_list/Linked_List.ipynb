# Linked List Pattern Sheet

## Recognition Rule

Linked-list questions are usually about preserving reachability while changing pointers. Before coding, state which nodes must remain reachable after every pointer update.

`Added bridge` marks a high-value problem added to complete a prerequisite ladder.

## 1. Single-Pass Traversal and Dummy Nodes

**Invariant:** `prev` is the tail of the already-correct list; `curr` is the first unprocessed node.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Delete Node in a Linked List (LC 237) | Local overwrite when predecessor is unavailable | Foundation |
| 2 | Remove Linked List Elements (LC 203) | Dummy head and conditional unlinking | Added bridge - must do |
| 3 | Remove Duplicates from Sorted List (LC 83) | Skip repeated successors | Must do |
| 4 | Remove Duplicates from Sorted List II (LC 82) | Remove an entire duplicate run | Must do |

## 2. Fast and Slow Pointers

**Invariant:** Fast advances twice as quickly, so slow identifies the midpoint when fast reaches the end.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Middle of Linked List (LC 876) | Learn the midpoint contract | Must do |
| 2 | Palindrome Linked List (LC 234) | Midpoint plus reverse-half comparison | Must do |
| 3 | Reorder List (LC 143) | Midpoint plus reverse plus alternating merge | Must do |

## 3. Cycle Detection

**Invariant:** If a cycle exists, fast and slow must meet; resetting one pointer converts the meeting offset into the distance to cycle entry.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Linked List Cycle (LC 141) | Detect a meeting | Must do |
| 2 | Linked List Cycle II (LC 142) | Locate the entry using Floyd's proof | Must do |

## 4. Fixed Gap and Pointer Alignment

| Order | Problem | Key invariant | Priority |
|---:|---|---|---|
| 1 | Remove Nth Node From End (LC 19) | Maintain a fixed gap of `n` nodes | Must do |
| 2 | Intersection of Two Linked Lists (LC 160) | Pointer switching equalizes path lengths | Must do |

## 5. Reversal and Reconnection

**Invariant:** `prev` is a fully reversed prefix, `curr` starts the untouched suffix, and `next` protects the suffix before rewiring.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Reverse Linked List (LC 206) | Reverse one complete chain | Must do |
| 2 | Swap Nodes in Pairs (LC 24) | Reverse fixed groups of two | Must do |
| 3 | Reverse Linked List II (LC 92) | Reverse one bounded segment and reconnect | Must do |
| 4 | Reverse Nodes in K-Group (LC 25) | Validate, reverse, and reconnect repeated groups | Advanced |
| 5 | Reverse Alternate K Nodes | Conditionally reverse alternating groups | Advanced variant |

## 6. Building and Merging Chains

**Invariant:** `tail` always points to the final node of the already-built sorted/output list.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Merge Two Sorted Lists (LC 21) | Merge two chains with a dummy head | Must do |
| 2 | Sort List (LC 148) | Split, recursively sort, then reuse two-list merge | Must do |
| 3 | Merge K Sorted Lists (LC 23) | Extend merge through divide-and-conquer or a heap | Must do |

## 7. Parallel Output Chains

**Invariant:** Each node is appended to exactly one stable partition; terminate every output tail before joining lists.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Odd Even Linked List (LC 328) | Maintain two position-based chains | Added bridge - recommended |
| 2 | Partition List (LC 86) | Maintain less-than and greater-or-equal chains | Must do |

## 8. Digit-by-Digit Arithmetic

**Invariant:** Each output digit is determined by aligned input digits and the incoming carry.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Add Two Numbers (LC 2) | Digits already arrive least-significant first | Must do |
| 2 | Add Two Numbers II (LC 445) | Restore reverse processing with stacks or reversal | Variant |
