Philosophy: Core Pattern -> Sub-pattern -> Problem ladder -> Bridge to the next problem. The goal is not to collect maximum problems. The goal is to recognize the idea, know the invariant/state, and modify a base template for variants.

## Labels

- MUST DO: high-return interview problem; do not skip if time is short.
- Variant: one important twist after the base problem.
- Advanced: solve after the base and variants are comfortable.
- Cross-link: useful connection, but the problem belongs mainly in another sheet.

## How to Study Each Sub-pattern

1. Solve the base problem slowly and write the invariant.
2. Read the bridge note before the next problem.
3. Change only one thing in the template: state, movement rule, answer update, or edge-case handling.
4. Dry run one tiny case and one boundary case.
5. Mark the mistake you made, not only whether the problem was accepted.

## Edge-Case Contract

Always check: empty input, one element, all equal elements, duplicates, negative values, overflow risk, impossible target, boundary indexes, and whether the answer needs exact condition or at-most condition.
