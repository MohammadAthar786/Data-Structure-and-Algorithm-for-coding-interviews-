# Matrix Pattern Sheet

## How to Use This Sheet

Study one block at a time. Write the invariant before coding, solve the base problem, and then solve each variant by changing one movement rule or state transition.

`Added bridge` marks an interview-worthy problem added to make the progression complete.

## 1. Direct Cell Traversal

**Invariant:** At iteration `(r, c)`, every previously visited cell has already contributed exactly once.

| Order | Problem | Focus | Priority |
|---:|---|---|---|
| 1 | Matrix Input and Output | Correct row/column indexing | Foundation |
| 2 | Add/Subtract Matrices | Independent element-wise update | Foundation |
| 3 | Transpose Matrix (LC 867) | Map `(r, c)` to `(c, r)` | Must do |
| 4 | Matrix Multiplication | Row-column dot product | Must do |

## 2. Boundary-State Traversal

**Invariant:** `top <= bottom` and `left <= right`; each completed edge is removed from the remaining rectangle.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Boundary Traversal | Visit four edges once | Foundation |
| 2 | Print Matrix in Wave/Zigzag | Alternate direction by row or column | Must do |
| 3 | Spiral Matrix (LC 54) | Shrink four boundaries after reading | Must do |
| 4 | Spiral Matrix II (LC 59) | Same invariant while writing | Variant |

## 3. Diagonal Invariants

**Invariant:** Main diagonal uses `r == c`; secondary diagonal uses `r + c == n - 1`; an anti-diagonal keeps `r + c` constant.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Matrix Diagonal Sum (LC 1572) | Recognize diagonal conditions | Must do |
| 2 | Print All Diagonals | Choose valid diagonal starting points | Must do |
| 3 | Diagonal Traverse (LC 498) | Preserve anti-diagonal while changing direction | Variant |

## 4. Linear Index Mapping

**Invariant:** For `cols` columns, `index = r * cols + c`, `r = index / cols`, and `c = index % cols`.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Reshape Matrix (LC 566) | Read and write through a linear index | Must do |
| 2 | Set Matrix Zeroes (LC 73) | Reuse first row/column as marker storage | Added bridge - must do |

## 5. Transpose-and-Reverse Transformations

**Invariant:** Rotation is a fixed composition of transpose and one axis reversal.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Rotate Image 90 Degrees Clockwise (LC 48) | Transpose, then reverse rows | Must do |
| 2 | Rotate 90 Degrees Anti-clockwise | Transpose, then reverse columns | Variant |

## 6. Search in Ordered Matrices

| Order | Problem | Key invariant | Priority |
|---:|---|---|---|
| 1 | Search a 2D Matrix (LC 74) | Treat globally sorted matrix as one sorted array | Added bridge - must do |
| 2 | Search a 2D Matrix II (LC 240) | From a corner, one comparison removes a row or column | Added bridge - must do |
| 3 | Kth Smallest Element in a Sorted Matrix (LC 378) | Count values `<= mid` monotonically | Added bridge - recommended |
| 4 | Matrix Median | Binary search a value whose matrix rank is monotonic | Added bridge - advanced |

## 7. Matrix Property Checks

| Order | Problem | Key invariant | Priority |
|---:|---|---|---|
| 1 | Identity Matrix Check | Diagonal is one; all other cells are zero | Foundation |
| 2 | Check Symmetric Matrix | `A[r][c] == A[c][r]` | Must do |
| 3 | Toeplitz Matrix (LC 766) | `A[r][c] == A[r-1][c-1]` | Must do |

## 8. Formula-Driven Generation

| Order | Problem | Key invariant | Priority |
|---:|---|---|---|
| 1 | Pascal Triangle (LC 118) | Current row depends on adjacent values above | Cross-link: DP/Math |
| 2 | Number Spiral (CSES) | Coordinates belong to a square layer | Recommended |

## Revision Order

`Transpose -> Spiral Matrix -> Diagonal Traverse -> Reshape -> Rotate Image -> Search Matrix I -> Search Matrix II -> Set Matrix Zeroes`
