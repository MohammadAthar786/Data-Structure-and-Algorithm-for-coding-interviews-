# DYNAMIC PROGRAMMING

<!-- DSA_SHEET_V2_LAYER -->

Philosophy: Core Pattern -> Sub-pattern -> Problem ladder -> Bridge to the next problem. The goal is not to collect maximum problems. The goal is to recognize the idea, know the invariant/state, and modify a base template for variants.

## Labels

- MUST DO: high-return interview problem; do not skip if time is short.
- Variant: one important twist after the base problem.
- Advanced: solve after the base and variants are comfortable.
- Cross-link: useful connection, but the problem belongs mainly in another sheet.

## How to Study Each Sub-pattern

1. Solve the base problem slowly and write the invariant.
2. Read the bridge note before the next problem.
3. Change only one thing in the template: state, movement rule, answer update, or edge-case handling.
4. Dry run one tiny case and one boundary case.
5. Mark the mistake you made, not only whether the problem was accepted.

## Edge-Case Contract

Always check: empty input, one element, all equal elements, duplicates, negative values, overflow risk, impossible target, boundary indexes, and whether the answer needs exact condition or at-most condition.

## Interview Execution Checklist

Use this before marking the pattern complete:

- Recognition: I can explain when this pattern applies and when it does not.
- Invariant: I know what condition/state must remain true during the loop or recursion.
- Template: I can write the base C++ skeleton without looking.
- Edge cases: I checked empty input, single element, duplicates, negative values, and boundary indexes where relevant.
- Complexity: I can justify time and space in one sentence.
- Practice ladder: I solved at least one easy, two medium, and one hard/variant problem for this pattern.

## Interview Answer Format

1. State the brute force idea and why it is slow.
2. Identify the pattern signal.
3. Explain the invariant or recurrence.
4. Walk through a small dry run.
5. Code cleanly.
6. Test edge cases.

## DP Conversion Checklist

For each DP problem, write these before coding:

- State: what does `dp[...]` mean exactly?
- Transition: how do smaller states build this state?
- Base case: what is known without recursion?
- Order: which direction fills states correctly?
- Answer: where is the final result stored?
- Optimization: can one dimension or previous rows be removed?
