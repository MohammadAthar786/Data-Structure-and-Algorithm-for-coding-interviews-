# Dynamic Programming: State-Transition Pattern Sheet

## DP Recognition Checklist

Use DP when recursive states repeat and the answer can be composed from smaller states. Define the state in one sentence, write the transition, identify base cases, and only then choose memoization or tabulation.

`Added bridge` marks a high-value prerequisite added to complete a ladder.

## 1. Linear One-Dimensional Recurrence

**Invariant:** `dp[i]` fully summarizes the answer for the prefix or value ending at state `i`.

| Order | Problem | State transition | Priority |
|---:|---|---|---|
| 1 | Climbing Stairs | Ways from the previous one or two states | Must do |
| 2 | Min Cost Climbing Stairs | Minimum cost from the previous one or two states | Must do |
| 3 | Friends Pairing Problem | Person stays single or pairs with one of `n-1` people | Recommended |
| 4 | Minimum/Maximum Path to One | Min/max over allowed predecessor states | Recommended |
| 5 | Decoding Ways (LC 91) | One-character and valid two-character transitions | Must do |

## 2. Index Skipping

**Invariant:** `dp[i]` is the best answer using positions up to `i`; taking `i` excludes a fixed neighborhood.

## Backtracking Cross-Links, Not DP Blocks

Restore IP Addresses and Expression Add Operators belong in Recursion/Backtracking unless a genuinely repeated state is memoized. Keep them as cross-links rather than presenting them as standard DP.
