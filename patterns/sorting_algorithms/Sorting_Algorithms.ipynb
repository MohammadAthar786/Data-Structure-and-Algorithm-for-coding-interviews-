# Sorting Algorithms

| Algorithm          | Time Complexity (Best / Avg / Worst) | Space    | Stable | Type             | When to Use                                 |
| ------------------ | ------------------------------------ | -------- | ------ | ---------------- | ------------------------------------------- |
| **Bubble Sort**    | O(n) / O(n) / O(n)                 | O(1)     |  Yes  | Comparison       | Educational only, rarely used               |
| **Selection Sort** | O(n) / O(n) / O(n)                | O(1)     |  No   | Comparison       | Simple but inefficient                      |
| **Insertion Sort** | O(n) / O(n) / O(n)                 | O(1)     |  Yes  | Comparison       | Nearly sorted arrays                        |
| **Merge Sort**     | O(n log n) / O(n log n) / O(n log n) | O(n)     |  Yes  | Divide & Conquer | Linked lists, stable sort needed            |
| **Quick Sort**     | O(n log n) / O(n log n) / O(n)      | O(log n) |  No   | Divide & Conquer | In-place, cache-friendly, fast average case |
| **Counting Sort**  | O(n + k)                             | O(k)     |  Yes  | Non-comp.        | Integers in small range                     |
