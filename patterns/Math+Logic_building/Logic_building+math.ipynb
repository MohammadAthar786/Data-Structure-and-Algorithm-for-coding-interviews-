# Logic Building and Math: Invariant-Based Foundation Sheet

## How to Use This Sheet

This is a prerequisite roadmap, not one algorithmic pattern. Work block by block and write the condition, loop invariant, or mathematical observation before implementation.

`Added bridge` marks an interview-worthy addition used to complete a progression.

**Invariant:** When processing prime `p`, mark multiples starting at `p*p`; smaller multiples were handled earlier.
