## 1.Monotonic Stack Pattern

Core Idea:
Maintain a stack in increasing or decreasing order to find the next or previous greater/smaller element efficiently.

1. Introductory Problems (Learn the Tool)

| Problem                            | Platform     | Difficulty | Core Idea                                       |
| ---------------------------------- | ------------ | ---------- | ----------------------------------------------- |
| Next Greater Element (NGE)         | LeetCode 496 | Easy       | Maintain decreasing stack, pop smaller elements |
| Next Smaller Element (NSE)         | Custom / GFG | Easy       | Maintain increasing stack, pop larger elements  |
| Previous Greater / Smaller Element | Custom / GFG | Easy       | Same as NGE/NSE but left side                   |

2. Application Level 1 (Direct Uses of NGE/NSE)

| Problem            | Platform     | Difficulty | Core Idea                                             |
| ------------------ | ------------ | ---------- | ----------------------------------------------------- |
| Daily Temperatures | LeetCode 739 | Medium     | NGE variant  track index difference                  |
| Stock Span Problem | GFG / Custom | Medium     | Previous Greater + distance                           |
| Remove K Digits    | LeetCode 402 | Medium     | Monotone increasing stack to maintain smallest number |
| 132 Pattern        | LeetCode 456 | Medium     | Reverse traversal + decreasing stack                  |
