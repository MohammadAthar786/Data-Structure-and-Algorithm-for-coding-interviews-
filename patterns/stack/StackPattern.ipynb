# Stack Pattern Sheet

## Recognition Rule

Use a stack when the newest unresolved item must be resolved first, or when a decision depends on the nearest unresolved item on one side.

`Added bridge` marks a high-return problem added to complete the ladder.

## 1. Fundamental LIFO State

| Order | Problem | Key invariant | Priority |
|---:|---|---|---|
| 1 | Valid Parentheses (LC 20) | Stack contains unmatched opening brackets | Added bridge - must do |
| 2 | Min Stack (LC 155) | Every state can recover its current minimum | Added bridge - must do |

## 2. Nearest Greater/Smaller Element

**Invariant:** Stack indices are unresolved and their values remain monotonic; a pop means the current value is their first valid answer.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Next Greater Element (LC 496) | Decreasing unresolved stack | Must do |
| 2 | Next Smaller Element | Reverse the comparison | Foundation |
| 3 | Previous Greater/Smaller Element | Change scan direction or answer side | Foundation |
| 4 | Daily Temperatures (LC 739) | Store indices to calculate distance | Must do |
| 5 | Stock Span Problem | Pop dominated history and aggregate span | Must do |

## 3. Span Boundaries and Contribution

**Invariant:** After popping, the stack top is the nearest boundary that prevents the current element from extending farther.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Largest Rectangle in Histogram (LC 84) | Find maximal width for each limiting height | Must do |
| 2 | Maximal Rectangle (LC 85) | Convert each matrix row into a histogram | Advanced |

## 4. Monotonic Greedy Construction

**Invariant:** The stack is the best lexicographically minimal feasible prefix after processing the current input prefix.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Remove K Digits (LC 402) | Pop harmful larger digits while removals remain | Must do |
| 2 | Remove Duplicate Letters (LC 316) | Add uniqueness and future-occurrence constraints | Added bridge - advanced |

## 5. Reverse-Scan Pattern Detection

| Order | Problem | Key invariant | Priority |
|---:|---|---|---|
| 1 | 132 Pattern (LC 456) | Reverse scan preserves candidates for `3` and the best popped `2` | Advanced |

## 6. Collision and Cancellation Simulation

**Invariant:** The stack stores survivors whose interaction with the current item is not yet settled.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Remove All Adjacent Duplicates (LC 1047) | Equal neighbors cancel | Cross-link from recursive variant |
| 2 | Asteroid Collision (LC 735) | Resolve repeated top-versus-current collisions | Must do |

## 7. Expression Conversion and Evaluation

**Invariant:** Operand order is preserved; operators leave the stack only when precedence and associativity permit.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Check Balanced Expression with Operators | Validate bracket and operator placement | Foundation |
| 2 | Postfix Evaluation | Evaluate an already precedence-free expression | Foundation |
| 3 | Evaluate Reverse Polish Notation (LC 150) | Handle signed integer operands | Must do |
| 4 | Infix to Postfix Conversion | Encode precedence with an operator stack | Must do |
| 5 | Infix to Prefix Conversion | Apply reverse-and-convert or direct handling | Optional |
| 6 | Basic Calculator II (LC 227) | Handle precedence without parentheses | Must do |
| 7 | Basic Calculator (LC 224) | Handle signs across nested parentheses | Must do |
| 8 | Decode String (LC 394) | Save outer count and partial result at each bracket | Added bridge - must do |
| 9 | Basic Calculator III (LC 772) | Combine nesting with full precedence | Advanced |

## 8. Recursion on a Stack

**Invariant:** Recursion temporarily owns removed elements; unwinding restores all elements except the intentional change.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Insert at Bottom | Learn remove-then-restore | Foundation |
| 2 | Reverse a Stack | Reuse insert-at-bottom | Must do |
| 3 | Find K-th Element from Top/Bottom | Track recursion depth | Optional |
| 4 | Delete an Element at Given Position | Skip one element at target depth | Foundation |
| 5 | Delete Middle Element of Stack | Compute target depth, then skip it | Must do |
| 6 | Check if Stack is Sorted | Compare while recursively exposing elements | Foundation |
| 7 | Insert in Sorted Order | Preserve sortedness while unwinding | Must do |
| 8 | Sort a Stack | Reuse sorted insertion | Must do |
| 9 | Remove Adjacent Duplicates Using Recursion | Cancellation variant; prefer iterative stack in interviews | Cross-link |

`LC 20 -> LC 155 -> NGE -> LC 739 -> Stock Span -> LC 84 -> LC 402 -> LC 735 -> LC 150 -> LC 227`
