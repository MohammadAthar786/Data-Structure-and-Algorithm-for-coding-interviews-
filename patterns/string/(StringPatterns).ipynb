# String Pattern Sheet

## How to Recognize the Pattern

Identify what must remain true while scanning: equal positions, token boundaries, carry, palindrome symmetry, prefix-suffix fallback, rolling hash, or trie prefix state.

`Added bridge` marks a problem introduced to complete a progression.

## 1. Positional Traversal and Comparison

**Invariant:** All characters or tokens before the current position have already compared equal.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Longest Common Prefix (LC 14) | Compare one character position across strings | Must do |
| 2 | Compare Version Numbers (LC 165) | Compare variable-length numeric tokens | Recommended |

## 2. Palindrome Expansion

**Invariant:** Before each expansion, the current interval is a palindrome; expand only while the next outer characters match.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Palindromic Substrings (LC 647) | Enumerate odd and even centers | Must do |
| 2 | Longest Palindromic Substring (LC 5) | Reuse expansion while tracking the best interval | Must do |

## 3. Rule-Based Scanning

| Order | Problem | Key invariant | Priority |
|---:|---|---|---|
| 1 | Roman to Integer (LC 13) | A smaller value before a larger one is subtractive | Must do |
| 2 | String to Integer - atoi (LC 8) | Parsing phases occur once: spaces, sign, digits, clamp | Must do |
| 3 | Valid Number (LC 65) | State captures which tokens remain legal | Added bridge - advanced |

## 4. Greedy String Construction

**Invariant:** Append the largest valid token that cannot make the remaining representation impossible.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Integer to Roman (LC 12) | Consume values from largest to smallest | Must do |

## 5. Arithmetic Simulation

**Invariant:** Every processed output position is final; unprocessed positions are represented by remaining digits and carry.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Add Binary (LC 67) | Right-to-left addition with carry | Must do |
| 2 | Add Strings (LC 415) | Generalize binary carry to decimal digits | Added bridge - recommended |
| 3 | Multiply Strings (LC 43) | Accumulate pairwise products by output position | Recommended |

## 6. Direct Pattern Matching

**Invariant:** At candidate start `i`, compare the pattern only against `text[i...]`.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Find the Index of First Occurrence (LC 28) | Brute-force matching at every start | Must do |

## 7. Prefix Function and KMP

**Invariant:** `lps[i]` is the longest proper prefix that is also a suffix ending at `i`; mismatch fallback never rechecks known matches.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Implement strStr - KMP Version (LC 28) | Build LPS, then search | Advanced |
| 2 | Repeated Substring Pattern (LC 459) | Interpret the final prefix-suffix length | Advanced |

## 8. Signature Hashing

**Invariant:** Strings belong to the same group exactly when their canonical signatures match.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Group Anagrams (LC 49) | Sorted or frequency-vector signature | Must do; cross-link: Hashmap |

## 9. Rolling Hash

**Invariant:** Update the next window hash by removing the outgoing contribution and adding the incoming character.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Repeated DNA Sequences (LC 187) | Fixed-size window fingerprints | Must do |
| 2 | Rabin-Karp String Search | Hash candidate windows, verify collisions | Advanced |

## 10. Trie and Prefix-State Search

**Invariant:** A trie node represents exactly the prefix consumed so far.

| Order | Problem | Progression | Priority |
|---:|---|---|---|
| 1 | Implement Trie (LC 208) | Insert, search, and starts-with | Must do |
| 2 | Design Add and Search Words (LC 211) | Wildcard creates DFS branches | Variant |
| 3 | Word Search II (LC 212) | Share prefixes while backtracking over a board | Advanced |
