# Level 4: String Simulation & Parsing

## Core Idea

These problems are primarily about:

* reading characters carefully
* applying rules
* simulating behavior
* parsing structured input

Unlike Sliding Window, Hashing, DP, or Two Pointers, there is usually no reusable state-maintenance template.

The challenge is correct implementation and edge-case handling.

---

## Recognition Signals

* convert one representation into another
* process characters sequentially
* apply special rules while scanning
* parse tokens/chunks
* build output according to a specification

---

## Sub-pattern 1: String Traversal & Comparison

### Core Idea

Compare strings character-by-character or position-by-position.

---

# Level 5: Advanced String Algorithms

## Core Idea

Used when brute-force substring matching becomes too expensive.

Focus on efficient pattern matching and string indexing.

---

## Recognition Signals

* repeated substring searching
* multiple pattern searches
* prefix/suffix relationships
* dictionary/prefix queries
* large string inputs

---



## Sub-pattern 1: Prefix Function / KMP

### Core Idea

Reuse information about previously matched prefixes.

Avoid restarting comparisons.

---

### Problem Ladder

| Order | Problem                             | Label    | Pattern         | Key Idea                       |
| ----- | ----------------------------------- | -------- | --------------- | ------------------------------ |
| 1     | Implement strStr() (KMP Version)    | Advanced | KMP             | Prefix-function based matching |
| 2     | Repeated Substring Pattern (LC 459) | Advanced | KMP Application | Prefix-suffix relationships    |

---

## Invariant

```text
LPS array stores
longest proper prefix
which is also suffix.
```

---

## Sub-pattern 2: Rolling Hash

### Core Idea

Represent substrings using hash values.

Compare hashes instead of comparing full substrings repeatedly.

---

### Problem Ladder

| Order | Problem                         | Label    | Pattern      | Key Idea                                |
| ----- | ------------------------------- | -------- | ------------ | --------------------------------------- |
| 1     | Repeated DNA Sequences (LC 187) | MUST DO  | Rolling Hash | Detect repeated fixed-length substrings |
| 2     | Rabin-Karp String Search        | Advanced | Rolling Hash | Pattern matching via hashes             |

---

## Invariant

```text
Update hash incrementally
when window moves.
```

---

## Sub-pattern 3: Trie / Prefix Matching

### Core Idea

Useful for dictionary and autocomplete-style problems.

---

### Problem Ladder
