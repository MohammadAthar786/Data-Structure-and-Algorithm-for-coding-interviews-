# Precomputation Techniques: Prefix & Suffix Arrays

## Core Philosophy

Precomputation means storing reusable information so future queries or computations become faster.

Instead of recomputing repeatedly:

```text
Build once → reuse many times
```

Main tradeoff:

| Benefit                     | Cost                |
| --------------------------- | ------------------- |
| Faster queries              | Extra preprocessing |
| Faster repeated computation | Extra memory        |

---

# Level 0: Running State

## Core Idea

Sometimes we do not need a full prefix/suffix array.

We only need to maintain one running value while scanning.

---

## Recognition Signals

* single pass problem
* only one final answer required
* no repeated range queries
* current answer depends on best/worst value seen so far

---

| Order | Problem                                          | Label   | Core Pattern           |
| ----- | ------------------------------------------------ | ------- | ---------------------- |
| 1     | Best Time to Buy and Sell Stock (LC 121)         | MUST DO | Running minimum        |
| 2     | Minimum Value to Get Positive Step Sum (LC 1413) | MUST DO | Running minimum prefix |

---

## Invariant

```text
Maintain only the information needed so far.
```

# Level 1: Basic Prefix and Suffix Arrays

## Core Idea

Store cumulative information from left-to-right or right-to-left.

---

## Prefix

Information accumulated from left → right.

Examples:

* prefix sum
* prefix max
* prefix xor
* prefix gcd

## Suffix

Information accumulated from right → left.

Examples:

* suffix sum
* suffix max
* suffix min

---

## Prefix Sum Formula

```text
prefix[i] = a[0] + a[1] + ... + a[i]
```

Range sum:

```text
sum(l, r) = prefix[r] - prefix[l - 1]
```

---

| Order | Problem                               | Label   | Core Pattern                 |
| ----- | ------------------------------------- | ------- | ---------------------------- |
| 1     | Range Sum Query - Immutable (LC 303)  | MUST DO | Prefix sum query             |
| 2     | Find Pivot Index (LC 724)             | MUST DO | Left sum = right sum         |
| 3     | Trapping Rain Water (LC 42)           | MUST DO | Prefix max + suffix max      |
| 4     | Product of Array Except Self (LC 238) | MUST DO | Left product + right product |

---

## Invariant

```text
Prefix stores information before or at index i.
Suffix stores information after or at index i.
```

# Level 2: Prefix Sum + HashMap Frequency

## Core Idea

Store frequency of previously seen prefix states.

Used mainly for counting subarrays.

---

## Most Important Rule

```text
Frequency map → count subarrays
```

---

## Universal Template

```python
freq = {0: 1}
prefix = 0
answer = 0

for x in arr:
    prefix += x

    answer += freq.get(prefix - target, 0)

    freq[prefix] = freq.get(prefix, 0) + 1
```

---

## Recognition Signals

* count subarrays
* exact sum
* divisible by k
* binary subarray count
* number of ranges satisfying equality

---

| Order | Problem                               | Label   | Core Pattern            |
| ----- | ------------------------------------- | ------- | ----------------------- |
| 1     | Subarray Sum Equals K (LC 560)        | MUST DO | Prefix sum + frequency  |
| 2     | Binary Subarrays With Sum (LC 930)    | MUST DO | Prefix sum + frequency  |
| 3     | Continuous Subarray Sum (LC 523)      | MUST DO | Prefix modulo           |
| 4     | Subarray Sums Divisible by K (LC 974) | MUST DO | Prefix modulo frequency |

---

## Invariant

```text
If current_prefix - previous_prefix = target,
then subarray sum is target.
```

# Level 3: Earliest Prefix Occurrence

## Core Idea

Instead of storing frequency, store the first index where a prefix state appeared.

Used mainly for longest subarray problems.

---

## Most Important Rule

```text
Earliest occurrence map → longest subarray
```

---

## Universal Template

```python
first_seen = {0: -1}
prefix = 0
best = 0

for i, x in enumerate(arr):
    prefix += x

    if prefix - target in first_seen:
        best = max(best, i - first_seen[prefix - target])

    if prefix not in first_seen:
        first_seen[prefix] = i
```

---

| Order | Problem                                     | Label    | Core Pattern                  |
| ----- | ------------------------------------------- | -------- | ----------------------------- |
| 1     | Maximum Size Subarray Sum Equals K (LC 325) | MUST DO  | Earliest prefix index         |
| 2     | Contiguous Array (LC 525)                   | MUST DO  | Convert 0 → -1, 1 → +1        |
| 3     | Longest Well-Performing Interval (LC 1124)  | Advanced | Prefix score + earliest state |

---

## Invariant

```text
To maximize length, preserve the earliest index of each prefix state.
```

# Level 4: Prefix XOR

## Core Idea

XOR has cancellation property.

```text
a XOR a = 0
```

Subarray XOR can be computed using prefix XOR.

---

## Formula

```text
subarray_xor(l, r) = prefixXor[r] XOR prefixXor[l - 1]
```

---

| Order | Problem                                          | Label   | Core Pattern            |
| ----- | ------------------------------------------------ | ------- | ----------------------- |
| 1     | Count Subarrays with Given XOR                   | MUST DO | Prefix XOR + hashmap    |
| 2     | Count Triplets That Can Form Equal XOR (LC 1442) | Variant | Equal prefix XOR states |

---

## Invariant

```text
If two prefix XOR values are equal,
the XOR between them is zero.
```

# Level 5: Difference Array

## Core Idea

Instead of updating every element in a range:

```text
Mark where effect starts.
Mark where effect ends.
Resolve later using prefix sum.
```

---

## Formula

```python
diff[l] += val
diff[r + 1] -= val
```

Final array is recovered using prefix sum.

---

## Recognition Signals

* multiple range updates
* add value to [l, r]
* bookings
* interval increments
* capacity changes over ranges

---

| Order | Problem                             | Label   | Core Pattern                   |
| ----- | ----------------------------------- | ------- | ------------------------------ |
| 1     | Range Addition (LC 370)             | MUST DO | Difference array               |
| 2     | Corporate Flight Bookings (LC 1109) | MUST DO | Range updates + prefix resolve |
| 3     | Car Pooling (LC 1094)               | MUST DO | Sweep-line difference array    |

---

## Invariant

```text
Difference array stores changes, not final values.
```

# Level 6: 2D Prefix Sum

## Core Idea

Extend prefix sum to matrices.

Used for rectangle sum queries.

---

## Formula

```text
prefix[i][j] = matrix[i][j] + top + left - topLeft
```

Rectangle query:

```text
sum = bottomRight - topExtra - leftExtra + overlap
```

---

| Order | Problem                                            | Label    | Core Pattern                 |
| ----- | -------------------------------------------------- | -------- | ---------------------------- |
| 1     | Range Sum Query 2D (LC 304)                        | MUST DO  | 2D prefix sum                |
| 2     | Matrix Block Sum (LC 1314)                         | MUST DO  | Rectangle query optimization |


---

## Invariant

```text
Use inclusion-exclusion to avoid double-counting matrix regions.
```

# Level 7: Contribution Technique

## Core Idea

Instead of generating all subarrays, calculate how much each element contributes.

Ask:

```text
In how many subarrays does this element contribute?
```

---

## Formula

```text
contribution = value × left_choices × right_choices
```

---

| Order | Problem                           | Label    | Core Pattern                        |
| ----- | --------------------------------- | -------- | ----------------------------------- |
| 1     | Sum of All Subarray Sums          | MUST DO  | Direct contribution count           |
| 2     | Sum of Subarray Minimums (LC 907) | Advanced | Monotonic stack + contribution      |
| 3     | Sum of Subarray Ranges (LC 2104)  | Advanced | Max contribution - min contribution |

---

## Invariant

```text
Count how many times each element participates instead of listing all ranges.
```

# Level 8: Prefix + Monotonic Data Structure

## Core Idea

Prefix hashmap works well for equality.

For inequalities and optimization, use monotonic structures.

---

## Recognition Signals

* shortest subarray
* at least k
* inequality condition
* negatives allowed
* first greater/smaller prefix
* optimal candidate among prefixes

---

| Order | Problem                                                | Label    | Core Pattern                   |
| ----- | ------------------------------------------------------ | -------- | ------------------------------ |
| 1     | Shortest Subarray with Sum at Least K (LC 862)         | Advanced | Prefix sum + monotonic deque   |
| 2     | Maximum Sum of Two Non-Overlapping Subarrays (LC 1031) | Variant  | Prefix best arrays             |
| 3     | Sum of Subarray Minimums (LC 907)                      | Advanced | Monotonic stack + contribution |

---

## Invariant

```text
Keep only prefix candidates that can still produce an optimal answer.
```

# Decision Tree

```text
Many static range sum queries?
    → Prefix sum array

Single-pass one final answer?
    → Running state

Count subarrays with exact property?
    → Prefix + hashmap frequency

Longest subarray with property?
    → Earliest prefix occurrence

Divisibility or remainder condition?
    → Prefix modulo

XOR subarray condition?
    → Prefix XOR

Need left and right information together?
    → Prefix/suffix arrays

Multiple range updates?
    → Difference array

2D rectangle sum query?
    → 2D prefix sum

Need shortest/optimal subarray with inequality?
    → Prefix + monotonic deque

Need total contribution over all subarrays?
    → Contribution technique
```

# Complexity Quick Reference

| Technique                | Build             | Query / Solve | Space  | Best Use Case            |
| ------------------------ | ----------------- | ------------- | ------ | ------------------------ |
| Running State            | O(n)              | O(n)          | O(1)   | Single final answer      |
| Prefix Sum               | O(n)              | O(1) query    | O(n)   | Many range queries       |
| Prefix + HashMap         | O(n)              | O(n) solve    | O(n)   | Count subarrays          |
| Earliest Prefix Map      | O(n)              | O(n) solve    | O(n)   | Longest subarray         |
| Prefix XOR               | O(n)              | O(n) solve    | O(n)   | XOR subarray problems    |
| Difference Array         | O(1) per update   | O(n) resolve  | O(n)   | Range updates            |
| 2D Prefix Sum            | O(mn)             | O(1) query    | O(mn)  | Matrix rectangle queries |
| Contribution             | O(n) / O(n log n) | O(n)          | varies | Count total effect       |
| Prefix + Monotonic Deque | O(n)              | O(n) solve    | O(n)   | Inequality optimization  |
